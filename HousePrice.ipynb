```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
#dependent Variable
Y=house4['SalePrice']
# independent Variable
X=house4['LotArea']
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```

```python
# The RSquare value is only .15. We need to add more variables to see if we can improve this. Let's try adding Total Bsmn SF
```

```python
#dependent Variable
Y=house4['SalePrice']
# independent Variable
X=house4[['LotArea', 'TotalBsmtSF']]
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```

```python
house4.info()
```

```python
# The R squared value is improved to .39. We still need to find better ways. Let's try adding ScreenPorch

#dependent Variable
Y=house4['SalePrice']
# independent Variable
X=house4[['LotArea', 'TotalBsmtSF','ScreenPorch']]
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```

```python
# can we add MSSubcalss

# The R squared value is improved to .39. We still need to find better ways. Let's try adding ScreenPorch

#dependent Variable
Y=house4['SalePrice']
# independent Variable
X=house4[['LotArea', 'TotalBsmtSF','ScreenPorch','MSSubClass']]
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```

```python
# the R squared value is now .43. Still not good enough
house5 = house4.dropna(subset =['LotFrontage'])
house5.info()
house4.info()
```

```python
# lets add lot frontage after cleaning the Nan values


#dependent Variable
Y=house5['SalePrice']
# independent Variable
X=house5[['LotArea', 'TotalBsmtSF','MSSubClass','LotFrontage']]
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```

```python
# the R Square is still around .46 only.
```

```python

# lets add MasVnRAre after cleaning the Nan values
house6 = house5.dropna(subset =['MasVnrArea'])
house6.info()


#dependent Variable
Y=house6['SalePrice']
# independent Variable
X=house6[['LotArea', 'TotalBsmtSF','MSSubClass','LotFrontage','MasVnrArea']]
#Adding constant                        
X=sm.add_constant(X)
simple=sm.OLS(Y,X)
result=simple.fit()
result.summary()
```
