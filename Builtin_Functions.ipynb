```python
# abs() method --- get absolute value of a number
#random integer
integer = -50
print ('Absolute value of integer is: ', abs(integer))
#random floating number
floating = -25.24
print ('Absolute value of floating is: ', abs(floating))
# any method --- the any() method returns True if any element of an iterable is true. If not, any() returns False
# The any() method takes an iterable (list, string, dictionary etc.) in Python.

# any() returns:

# True if at least one element of an iterable is true
# False if all elements are false or if an iterable is empty
# When	Return Value
#All values are true	True
#All values are false	False
#One value is true (others are false)	True
#One value is false (others are true)	True
#Empty Iterable	False

# with numbers
l=[1,2,3,4]
print('l is :', any(l))
l1 = [0, False]
print('l1 is :', any(l1))
l2 = [0, False,5]
print('l2 is :', any(l2))
lempty = []
print('lempty is :', any(lempty))
string = "this is good"
print('string is :', any(string))
zeroasvalue = [0]
print('zero as value is :', any(zeroasvalue))
zeroastext = '0'
print('zero as text is :', any(zeroastext))
```

```python
# all()

#The all() method returns True when all elements in the given iterable are true. If not, it returns False.
#The all() method takes a single parameter:

#iterable - any iterable (list, tuple, dictionary, etc.) which contains the elements

#How all() works for lists?

# all values true

print('l is :', all(l))
print('l1 is :', all(l1))
print('l2 is :', all(l2))
print('lempty is :', all(lempty))
print('string is :', all(string))
print('zero as value is :', all(zeroasvalue))
print('zero as text is :', all(zeroastext))
```

```python
# filter(), eval()


# The filter() method constructs an iterator from elements of an iterable for which a function returns true.
# How filter() works for iterable list?

# list of alphabets
alphabets = ['a', 'b', 'd', 'e', 'i', 'j', 'o']
# function that filters vowels
def filterVowels(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if(alphabet in vowels):
        return True
    else:
        return False

filteredVowels = filter(filterVowels, alphabets)

print('The filtered vowels are:')
for vowel in filteredVowels:
    print(vowel)

# eval()

#The eval() method parses the expression passed to this method and runs python expression (code) within the program.

x = 1

print('evaluation of x + 1 is', eval('x + 1'))

sentence = "testing strings to check if vowels can be filtered. Also, do the vowels get repeated in the filter?"
filteredVowelsfromsentence = filter(filterVowels, list(sentence))
print('The filtered vowels from sentence are:')
for vowel in filteredVowelsfromsentence:
    print(vowel)
```
